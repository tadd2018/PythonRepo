```python
import pandas as pd
df=pd.read_csv(r"C:\Users\tadde\Desktop\Excel Note\project form.csv")
print(df.describe())

```

```python

sumof=df['Sum of Daily hospital occupancy'].sum()
print('The sum of all dath',sumof)
```

```python
import matplotlib.pyplot as plt

# Sort by Sum of Daily hospital occupancy
df = df.sort_values('Sum of Daily hospital occupancy', ascending=False)

# Plot
df.plot(
    x='Entity',
    y='Sum of Daily hospital occupancy',
    kind='bar',
    color='ORANGE',
    figsize=(10, 6)
)

plt.title('Daily Hospital Occupancy by Country')
plt.xlabel('Entity')
plt.ylabel('Daily hospital occupancy')

# Save the chart to a file
plt.savefig('sales_report.png')
plt.show()
```
